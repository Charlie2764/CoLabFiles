**Problem 1:** Plot the function *y = x^3* for x=list(range(-5,6)). Plot this with squares at each
point and a dashed line connecting the squares. Make the line and the squares green. Make sure
the limits for the x and y axes are set to appropriate intervals. Set the font size of the numbers on
the x- and y-axes to 14. Add a title to the plot (name it whatever you want), and set the font size
of the title to be 18.

```python
import matplotlib.pyplot as plt
import numpy as np

x=list(range(-5,6))
y=list(range(-5,6))
i=0
while i < len(y):
  y[i] = x[i]**3
  i+=1
plt.plot(x,y, 'gs--')
plt.xlabel("years", fontsize = 14)
plt.ylabel("success", fontsize = 14)
plt.title("Progression", fontsize = 18)
```

```python
x2 = list(range(-10,11))
y2 = []
y3 = []
y4 = []
i = 0
while i < len(x2):
  if x2[i] == 0:
    y2.append(0)
  else:
    y2.append(1/(4*x2[i]))
  y3.append(np.exp(x2[i]/4))
  y4.append(np.sin(x2[i]/np.pi))
  i+=1
plt.subplot(3, 1, 1)
plt.plot(x2, y2, 'cx:')
plt.title("1/4x")
plt.subplot(3,1,2)
plt.plot(x2, y3, 'y^-.')
plt.title("e^(x/4)")
plt.subplot(3,1,3)
plt.plot(x2, y4, 'g.--')
plt.title("sin(x/10)")
plt.tight_layout()
```
